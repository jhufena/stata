## 2. effects

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Simulate some data
# Assume each composer has data for the first 35 years of life
years = np.arange(1, 36)
np.random.seed(5)  # Set random seed for reproducibility

# Simulating ICC data for Bach, Mozart, and Beethoven
# Ensuring the values are within the range [0, 1]
bach_icc = np.clip(np.random.normal(0.7, 0.05, 35), 0, 1)
mozart_icc = np.clip(np.random.normal(0.5, 0.05, 35), 0, 1)
beethoven_icc = np.clip(np.random.normal(0.6, 0.05, 35), 0, 1)

# Concatenate data into a single DataFrame
data = pd.DataFrame({
    'Year': np.concatenate([years, years, years]),
    'ICC': np.concatenate([bach_icc, mozart_icc, beethoven_icc]),
    'Composer': ['Bach']*35 + ['Mozart']*35 + ['Beethoven']*35
})

# Add some random slopes to simulate individual "dramatic arcs"
# Ensure the ICC values don't go beyond 0 or 1
data['ICC'] = np.clip(data['ICC'] + np.random.normal(0, 0.05, len(data)) * data['Year'] / 10, 0, 1)

# Scatter plot
sns.scatterplot(x='Year', y='ICC', hue='Composer', data=data, legend=None)  # Turn off the default legend

# Calculate and plot lines of best fit, and gather the line objects for the legend
line_objects = []
for composer in ['Bach', 'Mozart', 'Beethoven']:
    composer_data = data[data['Composer'] == composer]
    slope, intercept = np.polyfit(composer_data['Year'], composer_data['ICC'], 1)
    line, = plt.plot(composer_data['Year'], intercept + slope * composer_data['Year'], label=f"{composer} best fit")
    line_objects.append(line)

plt.title("Dramatic Arcs of Bach, Mozart, and Beethoven")
plt.xlabel("Year of Life")
plt.ylabel("Simulated ICC")

# Custom legend using the line objects
plt.legend(handles=line_objects, labels=['Bach best fit', 'Mozart best fit', 'Beethoven best fit'])
plt.show()
````
