# 3 Pre-Assessment

The pre-assessment is designed to guide students into the most appropriate course for their current skill level. Creating separate quizzes for each course level will help streamline this process.

## 3.1 Basic Stata Quiz

### Questions could include:

1. **True/False:** Stata is commonly used for qualitative data analysis.
2. **Multiple Choice:** Which command imports Excel data into Stata?
    - A) `read_excel`
    - B) `import excel`
    - C) `load_xl`
    - D) `xlread`
3. **Fill in the Blank:** The command to summarize data in Stata is `________`.
4. **Scenario Question:** You have a dataset with missing values. Which command would you use to eliminate these from your dataset?
5. **Code Interpretation:** What does the following Stata code do? `generate new_var = old_var * 2`

### Scoring:
- 0-2 Correct: Consider enrolling in the Introduction to Stata Programming course.
- 3-5 Correct: You may already possess the basics and might consider the Intermediate course.

## 3.2 Intermediate Stata Quiz

### Questions could include:

1. **True/False:** In Stata, a 'macro' is a place where you can store text or numbers for repeated use.
2. **Multiple Choice:** Which Stata command is used for merging datasets?
    - A) `concat`
    - B) `join`
    - C) `merge`
    - D) `attach`
3. **Code Interpretation:** Explain what the command `foreach var of varlist x1-x3 { replace `var' = `var' * 2 }` does.
4. **Scenario Question:** You need to create a new variable based on existing variables, how would you do this?
5. **Application:** What is the appropriate statistical method for comparing the means of three or more groups?

### Scoring:
- 0-2 Correct: Consider enrolling in the Intermediate Stata Programming course.
- 3-5 Correct: You may already be familiar with intermediate-level topics and might consider the Advanced course.
